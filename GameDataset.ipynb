```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\mbugu\OneDrive\Desktop\ML DATASET\Video Game Sales\vgsales.csv')

# Display the first few rows
print(df.head())

# Display basic information about the dataset
print(df.info())

# Display summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values in critical columns (e.g., 'Name', 'Year')
df.dropna(subset=['Name', 'Year'], inplace=True)

# Fill missing values in other columns with appropriate values (e.g., 0 for sales)
df['Publisher'].fillna('Unknown', inplace=True)

# Display the first few rows of the cleaned dataset
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate global sales by year
sales_by_year = df.groupby('Year')['Global_Sales'].sum().reset_index()

# Plot global sales trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_by_year, x='Year', y='Global_Sales', marker='o')
plt.title('Global Video Game Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
# Get the top 10 selling games
top_games = df.nlargest(10, 'Global_Sales')

# Plot top-selling games
plt.figure(figsize=(12, 6))
sns.barplot(data=top_games, x='Global_Sales', y='Name', palette='viridis')
plt.title('Top 10 Selling Video Games')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Game')
plt.show()
```

```python
# Sales by Genre
sales_by_genre = df.groupby('Genre')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_genre, x='Global_Sales', y='Genre', palette='viridis')
plt.title('Global Sales by Genre')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Genre')
plt.show()

# Sales by Platform
sales_by_platform = df.groupby('Platform')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_platform, x='Global_Sales', y='Platform', palette='viridis')
plt.title('Global Sales by Platform')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Platform')
plt.show()
```

```python
# Sales by Publisher
sales_by_publisher = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_publisher, x='Global_Sales', y='Publisher', palette='viridis')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Publisher')
plt.show()
```
